```python
import pandas as pd
dataset_filepath=r"\delaney_solubility_with_descriptors.csv"
dataset=pd.read_csv(dataset_filepath)
dataset.describe()
```

```python
y=dataset.logS
X=dataset.drop(['logS'],axis=1)
X.describe()
```

```python
from sklearn import ensemble
from sklearn.model_selection import train_test_split
from sklearn.metrics import r2_score,root_mean_squared_error
mod2 = ensemble.RandomForestRegressor(150,verbose=1,)
```

```python
mod2.fit(X,y,)
```

```python
y_=mod2.predict(X)
```

```python
print(r2_score(y,y_))
```

```python
print(root_mean_squared_error(y,y_))
```

```python
from sklearn.model_selection import train_test_split
_, val_x, _, val_y = train_test_split(X,y, random_state = 0)
```

```python
val_predictions = mod2.predict(val_x)
print(r2_score(val_y,val_predictions))
```
